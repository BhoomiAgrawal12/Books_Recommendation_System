### popularity based recommendation system,considering the highest average rating of the book voted by atleast 250 users

```python
ratingsWithBooks=ratings.merge(books,on="ISBN")
ratingsWithBooks.head()
```

```python
ratingsWithBooks.shape
```

```python
countNoOfRating_df=ratingsWithBooks.groupby('Book-Title').count()['Book-Rating'].reset_index()
countNoOfRating_df.rename(columns={'Book-Rating':'Count-Ratings'},inplace=True)
countNoOfRating_df
```

```python
averageRating_df=ratingsWithBooks.groupby('Book-Title').mean(numeric_only=True)['Book-Rating'].reset_index()
averageRating_df.rename(columns={'Book-Rating':'Avg-Rating'},inplace=True)
averageRating_df
```

```python
mostPopular_df=countNoOfRating_df.merge(averageRating_df,on='Book-Title')
mostPopular_df
```

```python
mostPopular_df=mostPopular_df[mostPopular_df['Count-Ratings']>=250].sort_values('Avg-Rating',ascending=False)
mostPopular_df
```

```python
mostPopular_df=mostPopular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','Count-Ratings','Avg-Rating']]
mostPopular_df
```

```python
mostPopular_df.head(50)
```

```python
x=ratingsWithBooks.groupby('User-ID').count()['Book-Rating']>200 #padhe likhe users boolean value tru or fase stored in X as list
padheUsers=x[x].index  #fiters the boolean serirs to keep only true val and index extract the index(user id) from the filtered series
padheUsers
```

```python
filteredRating=ratingsWithBooks[ratingsWithBooks['User-ID'].isin(padheUsers)]
```

```python
y = filteredRating.groupby('Book-Title').count()['Book-Rating']>=50
famousBooks = y[y].index
```

```python
final_ratings = filteredRating[filteredRating['Book-Title'].isin(famousBooks)]
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_scores = cosine_similarity(pt)
similarity_scores.shape
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```

```python
recommend('1984')
```
